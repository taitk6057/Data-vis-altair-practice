```python
# declarative visualization using altair (python vis method)
import pandas as pd   
import altair as alt 
alt.data_transformers.enable('json')
```

```python
url = "https://github.com/byuidatascience/data4python4ds/raw/master/data-raw/mpg/mpg.csv"
mpg = pd.read_csv(url)

#displ = cars engine size
#hwy = mpg on highway
```

```python
chart = (alt.Chart(mpg)
  .encode(
    x='displ', 
    y='hwy')
  .mark_circle()
)

chart.display()
```

```python
chart = (alt.Chart(mpg)
  .encode(
    x = "displ",
    y = "hwy",
    color = "class"
    )
  .mark_circle())

chart.display()
```

```python
#^^^^Sports cars have large engines like SUVs and pickup trucks, but 
#small bodies like midsize and compact cars, which improves their gas mileage

chartp = (alt.Chart(mpg)
  .encode(
    x = "displ",
    y = "hwy"
    )
  .mark_circle())

chartf = (alt.Chart(mpg)
  .encode(
    x = "displ",
    y = "hwy"
    )
  .transform_loess("displ", "hwy")
  .mark_line()
  )

chartp.display()
chartf.display()

#same plot different marks (geo object that represents the data)
```

```python
chartl = (alt.Chart(mpg)
  .transform_loess("displ", "hwy", groupby = ["drv"])
  .encode(
    x = "displ",
    y = "hwy",
    strokeDash = "drv"
    )
  .mark_line())

chartl.display()

#Here .mark_line() separates the cars into three lines based on their drv value, 
# which describes a carâ€™s drivetrain. One line describes all of the points with a 4 value, 
# one line describes all of the points with an f value, and one line describes all 
# of the points with an r value. Here, 4 stands for four-wheel drive, f for front-wheel drive, 
# and r for rear-wheel drive
```

```python
base = (alt.Chart(mpg.rename(columns = {"class": "class1"}))
  .encode(
    x = "displ",
    y = "hwy",
  ))

chart_smooth_sub = (base
  .transform_filter(alt.datum.class1 == "subcompact")
  .transform_loess("displ", "hwy")
  .mark_line()
)  

chart = base.encode(color = "class1").mark_circle() + chart_smooth_sub +chartl

chart.display()
```
